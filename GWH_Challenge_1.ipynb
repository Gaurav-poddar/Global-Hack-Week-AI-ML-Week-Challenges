```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#1. Bar chart

```python
# We will be showing population of top 5 cities in India through a bar graph
# DATA SOURCE = https://statisticstimes.com/demographics/country/india-cities-population.php

pop_x = pd.Series([31181377,20667655,14974073,12764935,11235018])
pop_y = pd.Series(['Delhi','Mumbai','Kolkata','Bangalore','Chennai'] )
pop_y
```

```python
fig = plt.figure(figsize = (12, 5))
plt.bar(pop_y,pop_x,color='Blue',width=0.5)
plt.xlabel("Name of Cities")
plt.ylabel("Population(ni Crores)")
plt.title("Population of Top 5 cities in India")
plt.show()
```

#2. Line Graph

```python
x = pd.Series([1,2,3,4])
y = pd.Series([5,6,7,8])
```

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(x,y)
plt.xlabel("X points")
plt.ylabel("Y points")
plt.title(" Straight line equation for y = x + 4 ")
plt.show()
```
